```python
import os
from langchain.schema import Document
from langchain_community.document_loaders import (
    LLMSherpaFileLoader,
    PDFMinerLoader,
    PDFPlumberLoader,
    PyPDFLoader,
)

RAW_DATA_DIR = "../raw_data/company/20240305"
FILENAME = "1709599445799.pdf"

file_path = os.path.join(RAW_DATA_DIR, FILENAME)
```

### 1. PyPDF

```python
%%time
loader = PyPDFLoader(file_path, extract_images=False)
raw_documents = loader.load()

print(f"# of pages: {len(raw_documents)}")
```

```python
print(raw_documents[0].page_content[368:1786] + "\n")
print("-" * 100)
print("\n" + raw_documents[2].page_content[33:])
```

### 2. PDFMiner

```python
loader = PDFMinerLoader(file_path)
raw_documents = loader.load()

print(f"# of pages: {len(raw_documents)}")
```

```python
print(raw_documents[0].page_content[720:2203] + "\n")
print("-" * 100)
print("\n" + raw_documents[0].page_content[4708:6045])
```

### 3. PDFPlumber

```python
loader = PDFPlumberLoader(file_path)
raw_documents = loader.load()

print(f"# of pages: {len(raw_documents)}")
```

```python
print(raw_documents[0].page_content[207:1872] + "\n")
print("-" * 100)
print("\n" + raw_documents[2].page_content[19:])
```

### 4. LLMSherpa

```python
%%time
loader = LLMSherpaFileLoader(file_path)
raw_documents = loader.load()

print(f"# of pages: {len(raw_documents)}")
```

```python
print("\n".join([raw_documents[i].page_content for i in range(1, 22)])[173:1825] + "\n")
print("-" * 100)
print("\n" + "\n".join([raw_documents[i].page_content for i in range(25, 34)]))
```
